```python
# -*- coding: utf-8 -*-
"""
Created on Sat Jan  6 18:46:32 2024

@author: Nisarg
"""

# Import the required libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import seaborn.objects as so

# Create a sample DataFrame
df = pd.DataFrame({
    'Day': ['Mon', 'Tue', 'Wed', 'Thur', 'Fri'],
    'Morning': [44, 46, 49, 59, 54],
    'Evening': [33, 46, 50, 49, 60]
})

# Set the index to Day column
df.set_index('Day', inplace=True)

# Create a stacked bar plot of Morning and Evening values
#sns.barplot(data=df, stacked=True)
Plot = so.Plot(df, x="Day", color="Morning").add(so.Bar(), so.Count(), so.Stack())

Plot.show()
# Show the plot
plt.show()
```
